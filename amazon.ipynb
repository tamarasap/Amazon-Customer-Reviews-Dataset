```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('../../Reviews.csv')
data.describe()
```

```python
data.head()
```

```python
data = data[data['Score'] != 3]

# napravi novu kolonu Sentiment
data['Sentiment'] = data['Score'].apply(lambda x: 'positive' if x > 3 else 'negative')


```

```python
print(data['Sentiment'].value_counts())
data.head()
```

```python
# Pronađi koliko imaš dostupnih negativnih (manjinska klasa)
min_class_size = data['Sentiment'].value_counts().min()

# Uzmi ili 5000 ili koliko god imaš ako je manje od toga
sample_n = min(15000, min_class_size)

negative = data[data['Sentiment'] == 'negative'].sample(n=sample_n, random_state=42)
positive = data[data['Sentiment'] == 'positive'].sample(n=sample_n, random_state=42)

data = pd.concat([negative, positive]).sample(frac=1, random_state=42)

print(data['Sentiment'].value_counts())
print(data.shape)
data.head()
```

```python
# sačuvaj kao CSV fajl
data.to_csv('Reviews_sample_balanced1.csv', index=False)

print("Fajl sacuvan kao Reviews_sample_balanced1.csv")
```
